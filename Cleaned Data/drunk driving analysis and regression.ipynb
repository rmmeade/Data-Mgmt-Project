```python
df = merged[['Crash_ID', 'Crash_Date', 'Prsn_Alc_Rslt_ID', 'Prsn_Injry_Sev_ID']]
```

```python
# check how many rows are not null
sum(df['Prsn_Alc_Rslt_ID'].notna())
```

```python
# check the percentage of missing data
100*(1-sum(df['Prsn_Alc_Rslt_ID'].notna())/len(df['Prsn_Alc_Rslt_ID']))
```

```python
import datetime as dt
df["Crash_Date"] = pd.to_datetime(df["Crash_Date"], format= "%m/%d/%Y")
```

```python
# divide the ridesharing periods
df["ridesharing"] = 'before'
df["ridesharing"].loc[(df["Crash_Date"] > dt.datetime(2014,6,1)) & (df["Crash_Date"] <= dt.datetime(2016,5,10))] = 'entry'
df["ridesharing"].loc[(df["Crash_Date"] > dt.datetime(2016,5,10)) & (df["Crash_Date"] <= dt.datetime(2017,5,29))] = 'exit'
df["ridesharing"].loc[(df["Crash_Date"] > dt.datetime(2017,5,29))] = 're-entry'
```

```python
# immputation
df.fillna(0,inplace = True)
```

```python
# define 'Y': drunk driving or not
def change(s):
    if int(s)==1: return 1
    else: return 0
df['Prsn_Alc_Rslt_ID'] = df['Prsn_Alc_Rslt_ID'].map(change)
```

```python
# regression
from patsy import dmatrices
x = dmatrices('Prsn_Alc_Rslt_ID ~ ridesharing', df, return_type='dataframe')[-1]
```

```python
y = df['Prsn_Alc_Rslt_ID']
```

```python
import statsmodels.api as sm
logit_model=sm.Logit(y, x)
result=logit_model.fit()
print(result.summary2())
```

```python
# visualization 
from pandas import Series
import matplotlib.pyplot as plt
ax = Series({'Intercept':-4.1111, 'entry':-0.0347,'exit':-0.1899,'re-entry':-0.3726}).plot(kind='bar',alpha=0.5)
plt.xticks(rotation=360)
ax.set_xlabel("ridesharing period: 'before' as base line")
ax.set_ylabel('coefficient')
plt.title('regression of drunk driving crashes on ridesharing period')
```
